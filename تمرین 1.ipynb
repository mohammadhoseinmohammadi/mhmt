```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python

X = np.random.rand(5) * 20
y = 2.358 * X - 3.121
y = y + np.random.normal(scale=3, size=y.shape)
X = X.reshape(-1, 1)
y = y.reshape(-1, 1)
plt.plot(X, y, 'bo')
lr_model = LinearRegression(normalize=True)


lr_model.fit(X, y)
y_line=lr_model.predict(X)
print(lr_model.score(X, y))
plt.plot(X[:, 0], y_line, 'r--')
```

```python
lr_model = LinearRegression(normalize=True)
X = np.random.rand(5) * 20
y = 2.358 * X - 3.121
y = y + np.random.normal(scale=3, size=y.shape)
X = X.reshape(-1, 1)
y = y.reshape(-1, 1)
plt.plot(X, y, 'bo')

x_line=np.arange(-10,30,0.1).reshape(-1, 1)
x_line = np.hstack((x_line, x_line ** 2, x_line ** 3, x_line ** 4))
lr_model.fit(x_line, y)

y_line=lr_model.predict(x_line)


print(lr_model.score(x_line, y_line))
plt.plot(x_line[:, 0], y_line, 'r--')
```
